```python
'''
(1) Your own median filter algorithm
'''
def my_median(image):
    _image = np.copy(image)
    for pixelX in range(1,image.shape[0]-2):
        for pixelY in range(1, image.shape[1]-2):
            mask = np.zeros((3,3))
            mask = image[pixelX-1:pixelX+1,pixelY-1:pixelY+1]
            _image[pixelX][pixelY] = np.median(mask, axis=None)
    return _image    
         
```

```python
'''
(2) skimage median filter
'''
import skimage

image = io.imread('tire.tif')
noise = random_noise(image,mode="s&p")
ski = skimage.filters.median(noise)
myImage = my_median(noise)
show_images([image,noise,ski,myImage],["original","noise","skimage","My Median"])
```

```python

'''
(3) Gaussion Filters
'''
image = io.imread('pout.tif')
noise = random_noise(image,mode="s&p")
gimage = skimage.filters.gaussian(noise)
gimage2 = skimage.filters.gaussian(noise,sigma=2)
gimage3 = skimage.filters.gaussian(noise,sigma=3)
show_images([noise,gimage,gimage2,gimage3],["original","sigma=1","sigma=2","sigma=3"])
```
